## Python program for Finding Amstrong or Not

```python
#Python program for Finding Amstrong or Not
number = int(input("Enter any number: "))
sum = 0
temp = number
while temp > 0:
   digit = temp % 10
   sum =sum + digit ** 3
   temp //= 10

if number == sum:
   print(number,"is an Armstrong number")
else:
   print(number,"is not an Armstrong number")
```
