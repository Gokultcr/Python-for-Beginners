## electricity bill calculation 

```python
#Program for calculating electricity bill
eleunit=int(input(" \nEnter the number of units consumed\n"))
if(eleunit<=50):
    Amountpayable=eleunit*0.05
elif(eleunit<=100):
    Amountpayable=(50*0.05)+(eleunit-50)*0.75
elif(eleunit<=200):
    Amountpayable=(50*0.05)+(100-50)*0.75+(eleunit-100)*1.20
else:
    Amountpayable=(50*0.05)+(100-50)*0.75+(200-100)*1.20+(eleunit-250)*1.50
    

print("\nElectricity bill=%.2f" %Amountpayable)
```
