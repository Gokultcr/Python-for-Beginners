## Program to print Prime numbers upto 50

```python
#Prime numbers upto 50
beg = 0
end = 50
  
for value in range(beg, end + 1): 
   if value > 1: 
       for n in range(2, value): 
           if (value % n) == 0: 
               break
       else: 
           print(value) 
```
